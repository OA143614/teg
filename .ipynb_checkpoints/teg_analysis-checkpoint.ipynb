```python
#sim_interval = 0.01
import random 
import numpy
numpy.set_printoptions(legacy='1.25')
list_index =[]

mean_low = 0
mean_high = 0
result = {}
numberInList = []
total_meanlow = 0
number_data = 60

#running to get the result
for i in range(number_data):
    ran_num = random.expovariate(1)
    result.update({i:ran_num})
    numberInList.append(i)


#finding mean_high

mean_high = (result[numberInList[0]] +result[numberInList[1]] +result[numberInList[2]]+result[numberInList[3]] +result[numberInList[4]])/5
print('meanhigh',mean_high)

#finding mean_low
for i in range(number_data):
    total_meanlow += result[numberInList[i]]
mean_low = total_meanlow/number_data
print('meanlow',mean_low)

#display the result
for i in range(number_data):
    if result[numberInList[i]] >= mean_high:
        print(numberInList[i],result[numberInList[i]],"high")
    elif result[numberInList[i]] <= mean_low:
        print(numberInList[i],result[numberInList[i]],"low")
        

```
